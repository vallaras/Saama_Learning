```python
from sklearn import datasets
import pandas as pd
```

```python
iris=datasets.load_iris()
iris_df=pd.DataFrame(iris.data)
iris_df['class']=iris.target
print(iris_df)
```

```python
iris_df.columns=['sepl_len','sepal_wind','petal_len','petal_wid','class']
```

```python
iris_df.dropna(how="all",inplace=True)
```

```python
iris_df
```

```python
iris_df["sepl_len"].dtype
```

```python
iris_df["sepl_len"]=iris_df["sepl_len"].astype('str')
```

```python
iris_df["sepal_wind"]=iris_df["sepal_wind"].astype('str')
```

```python
iris_df["petal_len"]=iris_df["petal_len"].astype('int')
```

```python
iris_df
```

```python
iris_df["petal_wid"].dtype
```

```python
listname=[]
for i in iris_df["petal_wid"]:
    if i>5:
        listname.append("big")
    else:
        listname.append("small")
```
