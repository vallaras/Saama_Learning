# iterating through Groups

```python
mark=pd.read_csv(r"C:\Users\ELCOT\Downloads\fortune1000.csv",index_col="Rank")
sec=mark.groupby("Secotor")
mark.head(3)
```

```python
#sec["Profits"].max()
df=pd.DataFrame(columns=mark.columns)
df
```

```python
for sect,data in sec:
    highst_revenue_company_in_group=data.nlargest(1,"Revenue")
    df=df.append(highst_revenue_company_in_group)
```
